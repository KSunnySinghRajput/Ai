# evalf() function
This function evaluates a given numerical expression upto a given floating point precision upto 100 digits. 

```python
from sympy.abc import r 
expr=pi*r**2 
expr
```

```python
expr.evalf(subs={r:5})
```

```python
expr=a/b 
expr.evalf(20, subs={a:100, b:3})
```

## Integral Transforms
1. SymPy supports various types of integral transforms as follows âˆ’

- laplace_transform
- fourier_transform
- sine_transform
- cosine_transform
- hankel_transform

```python
from sympy import fourier_transform, exp 
from sympy.abc import x, k 
expr=exp(-x**2) 
fourier_transform(expr, x, k)  #sqrt(pi)*exp(-pi**2*k**2)
```

```python
from sympy.integrals import laplace_transform 
from sympy.abc import t, s, a 
laplace_transform(t**a, t, s)
```
